```python
import os
import trimesh
import pandas as pd
import pickle
```

```python
vertex_counts = {}
fileList = os.listdir("../data/4-17/pulmonaryArterys_plys/")

for file in fileList:
    with open(f"../data/4-17/pulmonaryArterys_plys/{file}","rb") as f:
        shape = trimesh.exchange.ply.load_ply(f)
    # print(len(shape["vertices"]))
    vertex_counts[file] = len(shape["vertices"])

vertex_counts
```

```python
#保存为.pkl
with open("vertex_count.pkl","wb") as f:
    pickle.dump(vertex_counts,f)
```

```python
#查看点数最多和最少的文件
df1 = pd.DataFrame(vertex_counts,["vertex_num"])
ds1 = df1.loc["vertex_num"]

print(f"files with most vertices:\n{ds1.nlargest(10)}")
print("-"*30)
print(f"files with least vertices:\n{ds1.nsmallest(10)}")
```
